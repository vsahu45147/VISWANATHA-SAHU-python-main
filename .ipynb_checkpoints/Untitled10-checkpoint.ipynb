```python
import tkinter as tk
from tkinter import messagebox, ttk
import mysql.connector

# MySQL Connection
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="viswa@123klr",  # Replace with your MySQL password
    database="train_booking"
)
cursor = conn.cursor()

# GUI Window
root = tk.Tk()
root.title("Train Ticket Booking System")
root.geometry("600x500")

# Passenger Info Frame
tk.Label(root, text="Passenger Name").grid(row=0, column=0, padx=10, pady=5)
tk.Label(root, text="Age").grid(row=1, column=0, padx=10, pady=5)
tk.Label(root, text="Gender").grid(row=2, column=0, padx=10, pady=5)

name_var = tk.StringVar()
age_var = tk.StringVar()
gender_var = tk.StringVar()

tk.Entry(root, textvariable=name_var).grid(row=0, column=1)
tk.Entry(root, textvariable=age_var).grid(row=1, column=1)
ttk.Combobox(root, textvariable=gender_var, values=["Male", "Female", "Other"]).grid(row=2, column=1)

# Train Selection
tk.Label(root, text="Select Train").grid(row=3, column=0, padx=10, pady=5)
train_combo = ttk.Combobox(root)
cursor.execute("SELECT train_name FROM trains")
trains = [row[0] for row in cursor.fetchall()]
train_combo['values'] = trains
train_combo.grid(row=3, column=1)

# Booking Function
def book_ticket():
    name = name_var.get()
    age = age_var.get()
    gender = gender_var.get()
    train_name = train_combo.get()

    if not (name and age and gender and train_name):
        messagebox.showerror("Error", "Please fill all fields")
        return

    cursor.execute("SELECT train_id, fare FROM trains WHERE train_name=%s", (train_name,))
    train = cursor.fetchone()
    train_id, fare = train

    # Insert passenger
    cursor.execute("INSERT INTO passengers (name, age, gender) VALUES (%s, %s, %s)", (name, age, gender))
    passenger_id = cursor.lastrowid

    # Insert booking
    cursor.execute("INSERT INTO bookings (passenger_id, train_id, seat_no) VALUES (%s, %s, %s)",
                   (passenger_id, train_id, f"A{passenger_id}"))
    conn.commit()

    messagebox.showinfo("Success", f"Ticket booked successfully!\nFare: â‚¹{fare}")

# Button
tk.Button(root, text="Book Ticket", command=book_ticket, bg="green", fg="white").grid(row=4, column=1, pady=10)

# View Bookings Function
def view_bookings():
    view_window = tk.Toplevel(root)
    view_window.title("All Bookings")
    tree = ttk.Treeview(view_window, columns=("ID", "Name", "Train", "Seat", "Date"), show="headings")
    for col in ("ID", "Name", "Train", "Seat", "Date"):
        tree.heading(col, text=col)
    tree.pack(fill="both", expand=True)

    cursor.execute("""
        SELECT b.booking_id, p.name, t.train_name, b.seat_no, b.booking_date
        FROM bookings b
        JOIN passengers p ON b.passenger_id = p.passenger_id
        JOIN trains t ON b.train_id = t.train_id
    """)
    for row in cursor.fetchall():
        tree.insert("", "end", values=row)

tk.Button(root, text="View Bookings", command=view_bookings, bg="blue", fg="white").grid(row=5, column=1, pady=10)

root.mainloop()
```

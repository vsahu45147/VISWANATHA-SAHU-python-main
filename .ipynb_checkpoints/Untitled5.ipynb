```python
import random

def knight_random_walk():
    # Define the size of the chessboard
    board_size = 8
    # Define possible knight moves
    moves = [
        (2, 1), (1, 2), (-1, 2), (-2, 1),
        (-2, -1), (-1, -2), (1, -2), (2, -1)
    ]
    # Initialize chessboard with zeros
    board = [[0] * board_size for _ in range(board_size)]
    # Starting position of the knight
    x, y = random.randint(0, board_size - 1), random.randint(0, board_size - 1)
    board[x][y] = 1
    move_count = 1

    while True:
        # Get all valid moves
        valid_moves = []
        for dx, dy in moves:
            nx, ny = x + dx, y + dy
            if 0 <= nx < board_size and 0 <= ny < board_size and board[nx][ny] == 0:
                valid_moves.append((nx, ny))
        
        # If no valid moves, end the tour
        if not valid_moves:
            break

        # Randomly select a valid move
        x, y = random.choice(valid_moves)
        move_count += 1
        board[x][y] = move_count

    return board, move_count

# Run one random walk
board, steps = knight_random_walk()

# Display the final chessboard
for row in board:
    print(" ".join(f"{cell:2}" for cell in row))
print(f"\nThe knight completed {steps} steps.")
```

```python
def simulate_random_tours(num_tours=1_000_000):
    board_size = 8
    moves = [
        (2, 1), (1, 2), (-1, 2), (-2, 1),
        (-2, -1), (-1, -2), (1, -2), (2, -1)
    ]
    tour_lengths = [0] * (board_size * board_size + 1)

    for _ in range(num_tours):
        board = [[0] * board_size for _ in range(board_size)]
        x, y = random.randint(0, board_size - 1), random.randint(0, board_size - 1)
        board[x][y] = 1
        move_count = 1

        while True:
            valid_moves = []
            for dx, dy in moves:
                nx, ny = x + dx, y + dy
                if 0 <= nx < board_size and 0 <= ny < board_size and board[nx][ny] == 0:
                    valid_moves.append((nx, ny))
            
            if not valid_moves:
                break

            x, y = random.choice(valid_moves)
            move_count += 1
            board[x][y] = move_count
        
        tour_lengths[move_count] += 1

    return tour_lengths

# Run 1,000,000 random tours
tour_lengths = simulate_random_tours()

# Display the results
print("Tour Length | Frequency")
for length, freq in enumerate(tour_lengths):
    if freq > 0:
        print(f"{length:11} | {freq}")
```

```python
def find_full_tour():
    board_size = 8
    moves = [
        (2, 1), (1, 2), (-1, 2), (-2, 1),
        (-2, -1), (-1, -2), (1, -2), (2, -1)
    ]
    tour_lengths = [0] * (board_size * board_size + 1)
    attempts = 0

    while True:
        attempts += 1
        board = [[0] * board_size for _ in range(board_size)]
        x, y = random.randint(0, board_size - 1), random.randint(0, board_size - 1)
        board[x][y] = 1
        move_count = 1

        while True:
            valid_moves = []
            for dx, dy in moves:
                nx, ny = x + dx, y + dy
                if 0 <= nx < board_size and 0 <= ny < board_size and board[nx][ny] == 0:
                    valid_moves.append((nx, ny))
            
            if not valid_moves:
                break

            x, y = random.choice(valid_moves)
            move_count += 1
            board[x][y] = move_count
        
        tour_lengths[move_count] += 1

        if move_count == board_size * board_size:
            return attempts, tour_lengths

# Find the first full tour
attempts, tour_lengths = find_full_tour()

# Display the results
print(f"\nIt took {attempts} attempts to find the first full tour.")
print("Tour Length | Frequency")
for length, freq in enumerate(tour_lengths):
    if freq > 0:
        print(f"{length:11} | {freq}")
```

```python
import random
import time

# Board Size
BOARD_SIZE = 8

# Knight Moves
MOVES = [
    (2, 1), (1, 2), (-1, 2), (-2, 1),
    (-2, -1), (-1, -2), (1, -2), (2, -1)
]

# Accessibility Heuristic Function
def accessibility_heuristic():
    accessibility = [[8]*BOARD_SIZE for _ in range(BOARD_SIZE)]
    for x in range(BOARD_SIZE):
        for y in range(BOARD_SIZE):
            count = 0
            for dx, dy in MOVES:
                nx, ny = x + dx, y + dy
                if 0 <= nx < BOARD_SIZE and 0 <= ny < BOARD_SIZE:
                    count += 1
            accessibility[x][y] = count

    # Starting point
    x, y = random.randint(0, BOARD_SIZE - 1), random.randint(0, BOARD_SIZE - 1)
    board = [[0]*BOARD_SIZE for _ in range(BOARD_SIZE)]
    board[x][y] = 1
    move_count = 1

    while move_count < BOARD_SIZE * BOARD_SIZE:
        next_move = None
        min_access = float('inf')
        for dx, dy in MOVES:
            nx, ny = x + dx, y + dy
            if 0 <= nx < BOARD_SIZE and 0 <= ny < BOARD_SIZE and board[nx][ny] == 0:
                if accessibility[nx][ny] < min_access:
                    min_access = accessibility[nx][ny]
                    next_move = (nx, ny)

        if not next_move:
            break

        x, y = next_move
        board[x][y] = move_count = move_count + 1

    return board, move_count

# Brute Force Function
def brute_force():
    board = [[0] * BOARD_SIZE for _ in range(BOARD_SIZE)]
    x, y = random.randint(0, BOARD_SIZE - 1), random.randint(0, BOARD_SIZE - 1)
    board[x][y] = 1
    move_count = 1

    while move_count < BOARD_SIZE * BOARD_SIZE:
        valid_moves = []
        for dx, dy in MOVES:
            nx, ny = x + dx, y + dy
            if 0 <= nx < BOARD_SIZE and 0 <= ny < BOARD_SIZE and board[nx][ny] == 0:
                valid_moves.append((nx, ny))

        if not valid_moves:
            break

        x, y = random.choice(valid_moves)
        board[x][y] = move_count = move_count + 1

    return board, move_count

# Comparison Function
def compare_methods():
    # Run both methods
    start_time = time.time()
    _, heuristic_moves = accessibility_heuristic()
    heuristic_time = time.time() - start_time

    start_time = time.time()
    _, brute_moves = brute_force()
    brute_time = time.time() - start_time

    print("Comparison of Knight's Tour Approaches:")
    print("======================================")
    print(f"Accessibility Heuristic:")
    print(f"- Moves Taken: {heuristic_moves}")
    print(f"- Time Taken: {heuristic_time:.6f} seconds\n")
    
    print(f"Brute Force:")
    print(f"- Moves Taken: {brute_moves}")
    print(f"- Time Taken: {brute_time:.6f} seconds\n")

# Run Comparison
compare_methods()
```

```python
def format_array(array):
    """
    Reimplements NumPy-style formatting for 2D arrays containing positive integers.

    Parameters:
    array (list of lists): A 2D list containing positive integers.

    Returns:
    None
    """
    # Find the maximum element in the array
    max_width = max(len(str(element)) for row in array for element in row)
    
    # Format each row of the array
    formatted_rows = []
    for row in array:
        formatted_row = " ".join(f"{str(element).rjust(max_width)}" for element in row)
        formatted_rows.append(formatted_row)
    
    # Print the formatted rows
    for formatted_row in formatted_rows:
        print(formatted_row)


# Example usage
array = [
    [1, 234, 56789],
    [12, 3456, 78],
    [1234, 5, 6789]
]

print("Formatted 2D Array:")
format_array(array)
```

```python
def format_dataframe(data):
    # Calculate column widths (max of column values and header width)
    num_rows = len(data)
    num_cols = len(data[0]) if num_rows > 0 else 0
    column_widths = [max(len(str(data[row][col])) for row in range(num_rows)) for col in range(num_cols)]
    column_widths = [max(width, len(str(col))) for col, width in enumerate(column_widths)]  # Include column headers
    
    # Format the column headers
    header = "   " + " ".join(f"{str(col).rjust(column_widths[col])}" for col in range(num_cols))
    
    # Format the rows
    rows = []
    for row_idx, row in enumerate(data):
        formatted_row = f"{str(row_idx).rjust(2)} " + " ".join(
            f"{str(value).rjust(column_widths[col_idx])}" for col_idx, value in enumerate(row)
        )
        rows.append(formatted_row)
    
    # Print the DataFrame
    print(header)
    print("-" * len(header))
    for formatted_row in rows:
        print(formatted_row)


# Example usage
data = [
    [123, 45, 6789],
    [4, 5678, 9],
    [12, 3, 456]
]

print("Formatted DataFrame:")
format_dataframe(data)
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation

def quicksort(arr, start, end, bars, anim_frames):
    """
    Performs quicksort on the array and collects frames for animation.

    Parameters:
    arr (list): List of integers to sort.
    start (int): Starting index of the array/subarray.
    end (int): Ending index of the array/subarray.
    bars (list): List of bar heights for visualization.
    anim_frames (list): List to store frames for animation.
    """
    if start < end:
        pivot_index = partition(arr, start, end, bars, anim_frames)
        quicksort(arr, start, pivot_index - 1, bars, anim_frames)
        quicksort(arr, pivot_index + 1, end, bars, anim_frames)


def partition(arr, start, end, bars, anim_frames):
    """
    Partitions the array around a pivot and collects frames for animation.

    Parameters:
    arr (list): List of integers to partition.
    start (int): Starting index of the array/subarray.
    end (int): Ending index of the array/subarray.
    bars (list): List of bar heights for visualization.
    anim_frames (list): List to store frames for animation.
    """
    pivot = arr[end]
    i = start - 1

    for j in range(start, end):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
            bars[:] = arr[:]  # Update bar heights
            anim_frames.append(list(bars))  # Store frame for animation

    arr[i + 1], arr[end] = arr[end], arr[i + 1]
    bars[:] = arr[:]  # Update bar heights
    anim_frames.append(list(bars))  # Store frame for animation

    return i + 1


def animate_quicksort():
    """
    Creates and displays the animation of quicksort.
    """
    # Initial data for sorting
    data = [50, 20, 40, 80, 60, 10, 30, 70]
    bars = data[:]
    anim_frames = []

    # Perform quicksort and collect animation frames
    quicksort(data, 0, len(data) - 1, bars, anim_frames)

    # Initialize the plot
    fig, ax = plt.subplots()
    bar_container = ax.bar(range(len(data)), data, align="center")

    def update(frame):
        """Updates the bar heights for each frame."""
        for bar, height in zip(bar_container, frame):
            bar.set_height(height)
        return bar_container

    # Create the animation
    ani = animation.FuncAnimation(
        fig, update, frames=anim_frames, repeat=False, blit=False, interval=300
    )

    plt.title("Quicksort Animation")
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.show()


# Run the animation
animate_quicksort()
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation

def merge_sort(arr, left, right, bars, anim_frames):
    """
    Performs merge sort on the array and collects frames for animation.

    Parameters:
    arr (list): The array to sort.
    left (int): The starting index of the array/subarray.
    right (int): The ending index of the array/subarray.
    bars (list): Current state of the array for visualization.
    anim_frames (list): List to store frames for animation.
    """
    if left < right:
        mid = (left + right) // 2

        # Sort left half
        merge_sort(arr, left, mid, bars, anim_frames)

        # Sort right half
        merge_sort(arr, mid + 1, right, bars, anim_frames)

        # Merge sorted halves
        merge(arr, left, mid, right, bars, anim_frames)


def merge(arr, left, mid, right, bars, anim_frames):
    """
    Merges two sorted subarrays and collects frames for animation.

    Parameters:
    arr (list): The array to merge.
    left (int): Starting index of the left subarray.
    mid (int): Middle index, end of the left subarray.
    right (int): Ending index of the right subarray.
    bars (list): Current state of the array for visualization.
    anim_frames (list): List to store frames for animation.
    """
    left_part = arr[left:mid + 1]
    right_part = arr[mid + 1:right + 1]

    i, j, k = 0, 0, left

    while i < len(left_part) and j < len(right_part):
        if left_part[i] <= right_part[j]:
            arr[k] = left_part[i]
            i += 1
        else:
            arr[k] = right_part[j]
            j += 1
        k += 1
        bars[:] = arr[:]  # Update the visualization state
        anim_frames.append(list(bars))

    while i < len(left_part):
        arr[k] = left_part[i]
        i += 1
        k += 1
        bars[:] = arr[:]  # Update the visualization state
        anim_frames.append(list(bars))

    while j < len(right_part):
        arr[k] = right_part[j]
        j += 1
        k += 1
        bars[:] = arr[:]  # Update the visualization state
        anim_frames.append(list(bars))


def animate_merge_sort():
    """
    Creates and displays the animation of merge sort.
    """
    # Initial data for sorting
    data = [50, 20, 40, 80, 60, 10, 30, 70]
    bars = data[:]
    anim_frames = []

    # Perform merge sort and collect animation frames
    merge_sort(data, 0, len(data) - 1, bars, anim_frames)

    # Initialize the plot
    fig, ax = plt.subplots()
    bar_container = ax.bar(range(len(data)), data, align="center")

    def update(frame):
        """Updates the bar heights for each frame."""
        for bar, height in zip(bar_container, frame):
            bar.set_height(height)
        return bar_container

    # Create the animation
    ani = animation.FuncAnimation(
        fig, update, frames=anim_frames, repeat=False, blit=False, interval=300
    )

    plt.title("Merge Sort Animation")
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.show()


# Run the animation
animate_merge_sort()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

# Generate x values
x = np.linspace(0, 10, 400)

# Tricolor waves
y1 = np.sin(x)
y2 = np.sin(x) - 2

# Plot saffron and green waves
ax.fill_between(x, y1, color='#FF9933', alpha=0.8)
ax.fill_between(x, y2, -4, color='#138808', alpha=0.8)

# White middle
ax.fill_between(x, y1, y2, color='white')

# Ashoka Chakra
theta = np.linspace(0, 2*np.pi, 100)
r = 0.5
cx, cy = 5, -1

ax.plot(cx + r*np.cos(theta), cy + r*np.sin(theta),
        color='#000080', linewidth=2)

# Chakra spokes
for i in range(24):
    angle = i * (2*np.pi / 24)
    ax.plot([cx, cx + r*np.cos(angle)],
            [cy, cy + r*np.sin(angle)],
            color='#000080', linewidth=1)

# Text
ax.text(5, 1.5, "HAPPY", ha='center', fontsize=18, fontweight='bold')
ax.text(5, 1.0, "REPUBLIC DAY", ha='center',
        fontsize=26, fontweight='bold')
ax.text(5, 0.3, "26 January", ha='center',
        fontsize=16)

# Remove axes
ax.axis('off')

# Limits
ax.set_xlim(0, 10)
ax.set_ylim(-4, 3)

plt.show()
```

```python
import folium
from folium.plugins import MarkerCluster, Search

# Sample dataset: people and their coordinates
people = [
    {"name": "Viswa", "lat": 28.6139, "lon": 77.2090},    # Bhubaneswar
    {"name": "Nishant", "lat": 19.0760, "lon": 72.8777},  # Jagatsinghpur
    {"name": "Swati", "lat": 12.9716, "lon": 77.5946},    # Cuttack
    {"name": "Akankhya", "lat": 13.0827, "lon": 80.2707}, # Rourkela
]

# Create a map
m = folium.Map(location=[20, 77], zoom_start=5)

# Marker Cluster for better visualization
marker_cluster = MarkerCluster().add_to(m)

# Store marker references in a FeatureGroup (needed for search)
feature_group = folium.FeatureGroup(name="People").add_to(m)

# Dictionary to keep track of markers
marker_dict = {}

# Add markers
for person in people:
    marker = folium.Marker(
        location=[person['lat'], person['lon']],
        popup=person['name'],
        tooltip=person['name']
    )
    marker.add_to(marker_cluster)
    marker.add_to(feature_group)
    
    # Save marker reference in dictionary
    marker_dict[person['name']] = marker

# Add a search plugin
search = Search(
    layer=feature_group,
    search_label='tooltip',
    placeholder='Search for a person',
    collapsed=False
).add_to(m)

# Add JavaScript to open popup when search result is selected
m.get_root().html.add_child(folium.Element(f"""
<script>
    var markers = {{}};
    function registerMarkers() {{
        {"".join([f'markers["{name}"] = {{lat: {marker.location[0]}, lng: {marker.location[1]}}};' for name, marker in marker_dict.items()])}
    }}
    registerMarkers();
</script>
"""))

# Save map
m.save("viswa_map.html")
print("Map saved as viswa_map.html")
```

```python
import folium
from folium.plugins import Search

# Sample dataset
people = [
    {"name": "Viswa", "lat": 28.6139, "lon": 77.2090},    # Bhubaneswar
    {"name": "Nishant", "lat": 19.0760, "lon": 72.8777},  # Jagatsinghpur
    {"name": "Swati", "lat": 12.9716, "lon": 77.5946},    # Cuttack
    {"name": "Akankhya", "lat": 13.0827, "lon": 80.2707}, # Rourkela
]

# Create map
m = folium.Map(location=[20, 77], zoom_start=5)

# FeatureGroup for markers
feature_group = folium.FeatureGroup(name="People").add_to(m)

# Store marker references for JS
marker_js_list = []

# Add markers
for person in people:
    marker = folium.Marker(
        location=[person['lat'], person['lon']],
        popup=person['name'],
        tooltip=person['name']
    ).add_to(feature_group)
    
    marker_js_list.append({
        "name": person['name'],
        "lat": person['lat'],
        "lng": person['lon']
    })

# Add Search plugin
Search(
    layer=feature_group,
    search_label='tooltip',
    placeholder='Search for a person',
    collapsed=False
).add_to(m)

# Convert Python list to JS-friendly JSON string
marker_js_list_str = str(marker_js_list).replace("'", '"')

# Inject JS to open popup on search selection
m.get_root().html.add_child(folium.Element(f"""
<script>
    var marker_data = {marker_js_list_str};

    // Function to open popup by name
    function openPopupByName(name) {{
        for (var i=0; i<marker_data.length; i++) {{
            if(marker_data[i].name === name) {{
                var lat = marker_data[i].lat;
                var lng = marker_data[i].lng;
                map.setView([lat, lng], 10);
                var popup = L.popup()
                    .setLatLng([lat, lng])
                    .setContent(name)
                    .openOn(map);
                break;
            }}
        }}
    }}

    // Wait until the map is ready
    document.addEventListener('DOMContentLoaded', function() {{
        // Get search suggestions container
        var searchInputs = document.querySelectorAll('.leaflet-control-search input');
        searchInputs.forEach(function(searchBox) {{
            searchBox.addEventListener('input', function(e) {{
                var name = e.target.value;
                // Check if input matches any marker
                marker_data.forEach(function(marker) {{
                    if(marker.name.toLowerCase() === name.toLowerCase()) {{
                        openPopupByName(marker.name);
                    }}
                }});
            }});
        }});
    }});
</script>
"""))

# Save map
m.save("soa_map.html")
print("Map saved as soa_map.html")
```

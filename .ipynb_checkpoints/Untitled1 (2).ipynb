```python
import sqlite3
import sys

def create_table():
    conn = sqlite3.connect("C:\\Users\\hp\\Desktop\\pqr\\railway_reservation.db")
    cursor = conn.cursor()

    cursor.execute('''
    CREATE TABLE IF NOT EXISTS Tickets (
        ticket_id TEXT PRIMARY KEY,
        train_name TEXT,
        ticket_quantity INTEGER,
        ticket_price INTEGER) ''')

    conn.commit()
    conn.close()

def insert_tickets():
    conn = sqlite3.connect("C:\\Users\\hp\\Desktop\\pqr\\railway_reservation.db")
    cursor = conn.cursor()
    
    tickets_data = [
        ('T1', 'Train1', 3, 100),
        ('T2', 'Train1', 3, 150),
        ('T3', 'Train1', 3, 200),
        ('T4', 'Train1', 3, 250),
        ('T5', 'Train1', 3, 300)
    ]
    
    cursor.executemany('INSERT OR IGNORE INTO Tickets(ticket_id, train_name, ticket_quantity, ticket_price) VALUES (?, ?, ?, ?)', tickets_data)
    
    conn.commit()
    conn.close()

def get_tickets():
    conn = sqlite3.connect("C:\\Users\\hp\\Desktop\\pqr\\railway_reservation.db")
    cursor = conn.cursor()
    cursor.execute('SELECT * FROM Tickets')
    tickets = cursor.fetchall()
    conn.close()
    
    return tickets

def update_quantity(ticket_id, reserved_quantity):
    conn = sqlite3.connect("C:\\Users\\hp\\Desktop\\pqr\\railway_reservation.db")
    cursor = conn.cursor()
    cursor.execute('UPDATE Tickets SET ticket_quantity = ticket_quantity - ? WHERE ticket_id = ?', (reserved_quantity, ticket_id))
    conn.commit()
    conn.close()

# Running the functions
create_table()
insert_tickets()

# Print sys.prefix (as a debug statement)
print(sys.prefix)
```

```python
import pygame
import sys
import random

# Initialize Pygame
pygame.init()

# Set up the screen
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Cricket Game")

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)

# Fonts
font = pygame.font.SysFont(None, 40)

# Ball properties
ball_radius = 10
ball_speed = 5
ball_color = RED

# Bat properties
bat_width = 100
bat_height = 20
bat_color = WHITE

# Initialize ball position and velocity
ball_x = WIDTH // 2
ball_y = HEIGHT // 2
ball_dx = random.choice([-1, 1]) * ball_speed
ball_dy = random.choice([-1, 1]) * ball_speed

# Initialize bat position
bat_x = WIDTH // 2 - bat_width // 2
bat_y = HEIGHT - 50

# Game loop
clock = pygame.time.Clock()
score = 0
running = True
while running:
    screen.fill(BLACK)

    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Move the bat with arrow keys
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT]:
        bat_x -= 5
    if keys[pygame.K_RIGHT]:
        bat_x += 5

    # Bound the bat within the screen
    bat_x = max(0, min(bat_x, WIDTH - bat_width))

    # Move the ball
    ball_x += ball_dx
    ball_y += ball_dy

    # Check for collisions with walls
    if ball_x <= 0 or ball_x >= WIDTH - ball_radius:
        ball_dx *= -1
    if ball_y <= 0:
        ball_dy *= -1
    elif ball_y >= HEIGHT - ball_radius:
        # Check for collision with the bat
        if bat_x <= ball_x <= bat_x + bat_width:
            ball_dy *= -1
            score += 1
        else:
            # Game over
            text = font.render(f"Game Over! Score: {score}", True, WHITE)
            screen.blit(text, (WIDTH // 2 - text.get_width() // 2, HEIGHT // 2 - text.get_height() // 2))
            pygame.display.flip()
            pygame.time.wait(2000)  # Wait for 2 seconds before closing
            running = False

    # Draw the ball and the bat
    pygame.draw.circle(screen, ball_color, (ball_x, ball_y), ball_radius)
    pygame.draw.rect(screen, bat_color, (bat_x, bat_y, bat_width, bat_height))

    # Update the display
    pygame.display.flip()

    # Cap the frame rate
    clock.tick(60)

# Quit Pygame
pygame.quit()
sys.exit()
```

```python
import pygame
import sys
import random

# Initialize Pygame
pygame.init()

# Screen dimensions
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Cricket DRS System")

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
YELLOW = (255, 255, 0)

# Fonts
font = pygame.font.SysFont(None, 40)
small_font = pygame.font.SysFont(None, 30)

# Initialize ball and stumps position
ball_radius = 10
ball_x = WIDTH // 2
ball_y = 50

stump_width = 20
stump_height = 100
stump_x = WIDTH // 2 - stump_width // 2
stump_y = HEIGHT - 150

# Decision variables
decision_made = False
decision_text = ""

# Simulate ball tracking data (randomly generated for demo purposes)
def simulate_ball_tracking():
    impact = random.choice(["In Line", "Outside Off", "Outside Leg"])
    wickets = random.choice(["Hitting", "Missing"])
    return impact, wickets

def make_decision(impact, wickets):
    if impact == "In Line" and wickets == "Hitting":
        return "OUT"
    else:
        return "NOT OUT"

# Draw stumps
def draw_stumps():
    pygame.draw.rect(screen, WHITE, (stump_x, stump_y, stump_width, stump_height))
    pygame.draw.rect(screen, WHITE, (stump_x - 30, stump_y, stump_width, stump_height))
    pygame.draw.rect(screen, WHITE, (stump_x + 30, stump_y, stump_width, stump_height))

# Game loop
clock = pygame.time.Clock()
running = True
while running:
    screen.fill(BLACK)

    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN and not decision_made:
            if event.key == pygame.K_SPACE:
                # Simulate ball tracking when space is pressed
                impact, wickets = simulate_ball_tracking()
                decision_text = make_decision(impact, wickets)
                decision_made = True

    # Draw stumps
    draw_stumps()

    # Draw ball (moving down)
    if not decision_made:
        ball_y += 5

    pygame.draw.circle(screen, RED, (ball_x, ball_y), ball_radius)

    # Display decision if made
    if decision_made:
        text = font.render(f"Decision: {decision_text}", True, YELLOW if decision_text == "NOT OUT" else RED)
        impact_text = small_font.render(f"Impact: {impact}", True, WHITE)
        wickets_text = small_font.render(f"Wickets: {wickets}", True, WHITE)
        screen.blit(text, (WIDTH // 2 - text.get_width() // 2, HEIGHT // 2 - 50))
        screen.blit(impact_text, (50, HEIGHT - 100))
        screen.blit(wickets_text, (50, HEIGHT - 70))
    
    # Reset ball and decision after it goes off-screen
    if ball_y > HEIGHT and not decision_made:
        ball_y = 50

    # Update the display
    pygame.display.flip()

    # Cap the frame rate
    clock.tick(60)

# Quit Pygame
pygame.quit()
```

```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Set up the screen
screen_width, screen_height = 800, 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Simple Train Animation")

# Define colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
GRAY = (200, 200, 200)

# Create the track
track_y = screen_height // 2
track_height = 10

# Create the train
train_width, train_height = 60, 30
train_x = 0
train_y = track_y - train_height // 2
train_speed = 2

# Load a train image or use a simple rectangle if no image is available
# train_image = pygame.image.load('train.png')
train_image = pygame.Surface((train_width, train_height))
train_image.fill(GRAY)

# Main loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Move the train
    train_x += train_speed
    if train_x > screen_width:
        train_x = -train_width

    # Draw everything
    screen.fill(WHITE)
    pygame.draw.rect(screen, BLACK, (0, track_y, screen_width, track_height))
    screen.blit(train_image, (train_x, train_y))

    # Update the screen
    pygame.display.flip()

    # Control the frame rate
    pygame.time.Clock().tick(60)

# Quit Pygame
pygame.quit()
```

```python
import sqlite3

def create_table():
    conn = sqlite3.connect('C:\\Users\\hp\\Desktop\\pqr\\Reservation.db')
    cursor = conn.cursor()
    
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS Tickets (
            ticket_id TEXT PRIMARY KEY,
            movie_name TEXT,
            ticket_quantity INTEGER,
            ticket_price INTEGER)''')

    conn.commit()
    conn.close()
    
    def insert_Tickets():
        conn = sqlite3.connect('Reservation.db')
        cursor = conn.cursor()
    
    Tickets_data = [
            ('T1', 'Movie1', 3,50),
            ('T2', 'Movie2', 2,40),
            ('T3', 'Movie3', 4,60),
            ('T4', 'Movie4', 5,70),
            ('T5', 'Movie5', 1,65),
        ]
    
cursor.executemany('INSERT OR IGNORE INTO Tickets (ticket_id, movie_name, ticket_quantity,  ticket_price) VALUES (?, ?, ?, ?)', Tickets_data) # type: ignore

conn.commit() # type: ignore
conn.close() # type: ignore

def get_tickets():
     conn = sqlite3.connect('Reservation.db')
     cursor = conn.cursor()
     cursor.execute('SELECT * FROM Tickets')
     tickets = cursor.fetchall()
     conn.close()
     
     return tickets
 
def update_quantity(id,reserved_quantity):
     conn = sqlite3.connect('Reservation.db')
     cursor = conn.cursor()
     cursor.execute('UPDATE Tickets SET ticket_quantity = ticket_quantity - ? WHERE ticket_id = ?',(reserved_quantity,id))
     tickets = cursor.fetchall()
     conn.close()
     
create_table()
insert_Tickets() # type: ignore
import customtkinter
from tkinter import *
from tkinter import ttk
import tkinter as tk
from tkinter import messagebox
import database

app = customtkinter.CTk()
app.title('Tickets Booking System')
app.geometry('600x600')
app.config(bg='#18161D')
app.resizable(False,False)

font1 = ('Arial',25,'bold')
font2 = ('Arial',13,'bold') 
font3 = ('Arial',18,'bold')  

def add_to_treeview():
    tickets = database.get_tickets()
    tree.delete(*tree.get_children())
    for tickets in tickets:
        if tickets[2] > 0:
            tree.insert('', END, values=tickets)
 
def reservation(name, movie, quantity, price):
     customer_name = name
     movie_name = movie
     booked_quantity = quantity
     ticket_price = price
     total_price = ticket_price * booked_quantity
     
     frame = customtkinter.CTkFrame(app,bg_color='#18161D',fg_color='#292933',corner_radius=10,border_width=2,border_color='#0f0',width=200,height=130)
     frame.place(X=190,y=450)
     
     name_label = customtkinter.CTkLabel(frame,font=font3,text=f'Name:{customer_name}',text_color='#fff',bg_color='#292933')
     name_label.place(x=10,y=10)
     
     movie_label = customtkinter.CTkLabel(frame,font=font3,text=f'Movie:{movie_name}',text_color='#fff',bg_color='#292933')
     movie_label.place(x=10,y=50)
     
     total_price_label = customtkinter.CTkLabel(frame,font=font3,text=f'Total Price:{total_price}',text_color='#fff',bg_color='#292933')
     total_price_label.place(x=10,y=90)
     
     return total_price
 
def book():
     customer_name = name_entry.get()
     selected_item = tree.focus()
     if not selected_item:
         messagebox.showerror('Error', 'Choose a ticket to book. ')
     elif not customer_name:
        messagebox.showerror('Error', 'Enter customer name.')
     else:
         row = tree.item(selected_item)['values']
         movie_name = row[1]
         ticket_price = row[3]
         booked_quantity = int(variable.get()) # type: ignore
         if booked_quantity > row[2]:
             messagebox.showerror('Error', 'No enough tickets. ')
         else:
              database.update_quantity(row[0],booked_quantity)
              add_to_treeview()
              total_price = reservation(customer_name, movie_name, booked_quantity, ticket_price)
              with open('Tickets.txt', 'a') as file:
                file.write(f'Customer Name: {customer_name}\n')
                file.write(f'Movie Name: {movie_name}\n')
                file.write(f'Total Price: {total_price}$\n=====================\n')
              messagebox.showinfo('Success', 'Tickets are booked!')
    
title1_label = customtkinter.CTkLabel(app,font=font1,text='Available Films',text_color='#fff',bg_color='#18161D')
title1_label.place(x=200,y=20)

image1 = PhotoImage(file="1.png")
image1_label = Label(app, image=image1,bg='#18161D')
image1_label.place(x=130,y=10)

image2 = PhotoImage(file="2.png")
image2_label = Label(app, image=image2,bg='#18161D')
image2_label.place(x=380,y=5)

name_label = customtkinter.CTkLabel(app,font=font3,text='Customer Name:',text_color='#fff',bg_color='#18161D')
name_label.place(x=120,y=300)

name_entry = customtkinter.CTkEntry(app,font=font3,text_color='#000',fg_color='#fff',border_color='#AA04A7',border_width=2,width=160)
name_entry.place(x=290,y=300)

number_label = customtkinter.CTkLabel(app,font=font3,text='No. of Tickets:',text_color='#fff',bg_color='#18161D')
number_label.place(x=122,y=350)

variable1 = StringVar()
options = ['1','2','3']

duration_options = customtkinter.CTkComboBox(app,font=font3,text_color='#000',button_color='#AA04A7',button_hover_color='#AA04A7',border_color='#AA04A7',width=160,variable=variable1,values=options,state='readonly')
duration_options.set('1')
duration_options.place(x=290,y=350)

book_button = customtkinter.CTkButton(app,font=font3,text_color='#fff',text='Book Tickets',fg_color='#AA04A7',hover_color='#6D006B',bg_color='#18161D',cursor='hand2',corner_radius=15,width=200)
book_button.place(x=190,y=400)

style = ttk.Style(app)

style.theme_use('clam')
style.configure('Treeview',font=font2,foreground='#fff',background='#000',fieldbackground='#292933')
style.map('Treeview',background=[('selected', '#AA04A7')])

tree = ttk.Treeview(app,height=8)

tree['columns'] = ('Ticket ID', 'Movie Name', 'Avilable Tickets', 'Ticket Price')

tree.column('#0',width=0,stretch=tk.NO)
tree.column('Ticket ID', anchor=tk.CENTER, width=100,)
tree.column('Movie Name', anchor=tk.CENTER, width=100,)
tree.column('Avilable Tickets', anchor=tk.CENTER, width=100,)
tree.column('Ticket Price', anchor=tk.CENTER, width=100,)

tree.heading('Ticket ID', text='Ticket ID')
tree.heading('Movie Name', text='Movie Name')
tree.heading('Avilable Tickets', text='Avilable Tickets')
tree.heading('Ticket Price', text='Ticket Price')

tree.place(x=90,y=95)

add_to_treeview()

app.mainloop()
```

```python
df.columns = df.columns.str.strip()
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

scaler = MinMaxScaler()

array = []
train_data = []
train_labels = []

# Verify the column name 'Milk Production'
if 'Milk Production' in df.columns:
    for i in range(len(df)):
        array.append(df.iloc[i]['Milk Production'])

    array = np.array(array).reshape(-1, 1)
    array = scaler.fit_transform(array)

else:
    print("Column 'Milk Production' not found in the DataFrame.")
```

```python
k = 0
for i in range(len(array)):
    try:
        train_data.append(array[12*k:12*(k+1)])
        train_labels.append(array[12*(k+1)])
        k+=1
    except:
        break

train_data = np.squeeze(train_data)
train_labels = np.array(train_labels)

train_data.shape
```

```python
train_data = train_data[:len(train_labels)]
train_data = np.expand_dims(train_data,1)
train_data.shape
```

```python
train_data[1]
```

```python
train_labels[1]
```

```python
model = Sequential()

model.add(LSTM(250,input_shape=(1,12)))
model.add(Dropout(0.5))


model.add(Dense(250,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(12,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1,activation='relu'))

model.compile(loss='mean_squared_error',optimizer='adam')

model.summary()
```

```python
E = 1000

H = model.fit(train_data,train_labels,epochs=E)
```

```python
epochs = range(0,E)
loss = H.history['loss']
plt.plot(epochs,loss)
```

```python
preds = scaler.inverse_transform(model.predict(train_data))
plt.plot(range(0,13),preds,label='our predictions')
plt.plot(range(0,13),scaler.inverse_transform(train_labels),label='real values')
plt.legend()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import make_blobs 
from sklearn.svm import SVC 
## Create synthetic customer data 
X, y = make_blobs(n_samples=100, centers=2, random_state=6,
cluster_std=2.5) 
## Initialize an SVM model
model = SVC(kernel='linear') 
model.fit(X, y)  
## Define function to plot decision boundary 
def plot_decision_boundary(X, y, model):
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='autumn') 
    ax = plt.gca() 
    xlim = ax.get_xlim()  
    ylim = ax.get_ylim()
    ## Create grid to evaluate the model
    xx = np.linspace(xlim[0], xlim[1], 30) 
    yy = np.linspace(ylim[0], ylim[1], 30)  
    YY, XX = np.meshgrid(yy, xx)  
    xy = np.vstack([XX.ravel(), YY.ravel()]).T  
    Z = model.decision_function(xy).reshape(XX.shape)
    ## Plot decision boundary and margins
    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
linestyles=['--', '-', '--']) 
    ## Plot support vectors 
    ax.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
s=100, linewidth=1, facecolors='none', edgecolors='k')  
    plt.title('SVM Classification Decision Boundary')  
    plt.xlabel('Customer Age') 
    plt.ylabel('Customer Income')  
    plt.show()  
## Call the plot function 
plot_decision_boundary(X, y, model)
```

```python
def encode_message(message):
    binary = ''.join(format(ord(char), '08b') for char in message)
    return binary

def decode_message(binary):
    chars = [binary[i:i+8] for i in range(0, len(binary), 8)]
    message = ''.join(chr(int(char, 2)) for char in chars)
    return message

# Example usage
message = "Laser Test"
binary_code = encode_message(message)
print("Encoded:", binary_code)
decoded_message = decode_message(binary_code)
print("Decoded:", decoded_message)
```

```python
import matplotlib.pyplot as plt

# Constants
current_population = 8.1e9  # Current world population in billions
growth_rate = 0.009  # Annual growth rate as a decimal (0.9%)

# Initialize variables
population = current_population
years = []
populations = []
annual_increases = []
population_doubling_year = None
population_quadrupling_year = None

# Simulate 100 years of growth
for year in range(1, 101):
    annual_increase = population * growth_rate
    new_population = population + annual_increase
    
    # Append data for visualization
    years.append(year)
    populations.append(new_population / 1e9)  # Convert to billions for readability
    annual_increases.append(annual_increase / 1e9)
    
    # Check for doubling and quadrupling milestones
    if not population_doubling_year and new_population >= 2 * current_population:
        population_doubling_year = year
    if not population_quadrupling_year and new_population >= 4 * current_population:
        population_quadrupling_year = year
    
    # Update population for the next year
    population = new_population

# Plotting
plt.figure(figsize=(12, 6))

# Population over time
plt.plot(years, populations, label="World Population (billions)", color="blue", linewidth=2)

# Milestone annotations
if population_doubling_year:
    plt.axvline(population_doubling_year, color="green", linestyle="--", label=f"Doubling Year: {population_doubling_year}")
if population_quadrupling_year:
    plt.axvline(population_quadrupling_year, color="red", linestyle="--", label=f"Quadrupling Year: {population_quadrupling_year}")

# Add titles and labels
plt.title("World Population Growth Over 100 Years (Assuming Constant Growth Rate)", fontsize=14)
plt.xlabel("Years", fontsize=12)
plt.ylabel("Population (billions)", fontsize=12)

# Add legend
plt.legend()

# Grid and show the plot
plt.grid(alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

def flip_coin(num_flips):
    """Simulate flipping a coin."""
    results = [random.choice([1, 2]) for _ in range(num_flips)]
    heads = results.count(1)
    tails = results.count(2)
    return heads, tails

def plot_results(num_flips, heads, tails):
    """Plot the results of the coin flips with frequencies and percentages."""
    total = heads + tails
    percentages = [heads / total * 100, tails / total * 100]
    labels = ['Heads', 'Tails']
    frequencies = [heads, tails]

    plt.figure(figsize=(8, 5))
    bars = plt.bar(labels, frequencies, color=['blue', 'orange'], alpha=0.7)
    
    # Add text on top of bars for frequencies and percentages
    for bar, freq, perc in zip(bars, frequencies, percentages):
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                 f"{freq}\n({perc:.2f}%)", ha='center', va='bottom', fontsize=10)
    
    # Chart labels and title
    plt.title(f"Coin Flip Simulation for {num_flips} Flips", fontsize=14)
    plt.ylabel("Frequency", fontsize=12)
    plt.ylim(0, max(frequencies) * 1.1)
    plt.grid(axis='y', alpha=0.4)
    plt.tight_layout()
    plt.show()

# Simulate and plot for different numbers of flips
for num_flips in [200, 20_000, 200_000]:
    heads, tails = flip_coin(num_flips)
    print(f"\nResults for {num_flips} flips:")
    print(f"Heads: {heads} ({heads / num_flips * 100:.2f}%)")
    print(f"Tails: {tails} ({tails / num_flips * 100:.2f}%)")
    plot_results(num_flips, heads, tails)
```

```python
import random
import sys
import seaborn as sns
import matplotlib.pyplot as plt

def roll_two_dice(num_rolls):
    """Simulate rolling two dice and return the frequencies of sums."""
    sums = [random.randint(1, 6) + random.randint(1, 6) for _ in range(num_rolls)]
    frequencies = {i: sums.count(i) for i in range(2, 13)}
    return frequencies

def plot_results(frequencies, num_rolls):
    """Plot the results of rolling two dice using a horizontal bar plot."""
    sums = list(frequencies.keys())
    counts = list(frequencies.values())
    percentages = [count / num_rolls * 100 for count in counts]

    plt.figure(figsize=(10, 6))
    sns.barplot(x=percentages, y=sums, orient='h', palette='viridis')
    
    # Add labels and title
    plt.title(f"Distribution of Sums for {num_rolls:,} Rolls of Two Dice", fontsize=14)
    plt.xlabel("Percentage (%)", fontsize=12)
    plt.ylabel("Sum of Two Dice", fontsize=12)
    plt.grid(axis='x', alpha=0.4)
    
    # Display percentage labels
    for i, (percent, count) in enumerate(zip(percentages, counts)):
        plt.text(percent, i, f"{count} ({percent:.2f}%)", va='center', fontsize=10)
    
    plt.tight_layout()
    plt.show()

def main():
    """Main function to handle command-line input and simulation."""
    # Check for command-line argument
    if len(sys.argv) != 2:
        print("Usage: python roll_two_dice.py <number_of_rolls>")
        return

    # Get the number of rolls from the command line
    try:
        num_rolls = int(sys.argv[1])
        if num_rolls <= 0:
            raise ValueError
    except ValueError:
        print("Please provide a positive integer for the number of rolls.")
        return

    # Simulate rolling two dice
    frequencies = roll_two_dice(num_rolls)

    # Plot the results
    plot_results(frequencies, num_rolls)

# Run the program
if __name__ == "__main__":
    main()
```

```python
def display_board(board):
    """Display the current Tic-Tac-Toe board."""
    print("\nCurrent board:")
    for row in board:
        print(" | ".join(row))
        print("-" * 5)

def check_winner(board):
    """Check if there's a winner or the game is a draw."""
    # Check rows, columns, and diagonals
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != " ":
            return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] != " ":
            return board[0][i]
    if board[0][0] == board[1][1] == board[2][2] != " ":
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != " ":
        return board[0][2]
    # Check for a draw
    if all(cell != " " for row in board for cell in row):
        return "Draw"
    return None

def two_player_tic_tac_toe():
    """Main function to play two-player Tic-Tac-Toe."""
    board = [[" " for _ in range(3)] for _ in range(3)]
    players = ["X", "O"]
    turn = 0

    while True:
        display_board(board)
        current_player = players[turn % 2]
        print(f"Player {current_player}, it's your turn.")

        # Input validation
        while True:
            try:
                row, col = map(int, input("Enter row and column (0-2) separated by a space: ").split())
                if board[row][col] == " ":
                    board[row][col] = current_player
                    break
                else:
                    print("That square is already taken. Try again.")
            except (ValueError, IndexError):
                print("Invalid input. Enter row and column numbers between 0 and 2.")

        # Check for a winner or draw
        result = check_winner(board)
        if result:
            display_board(board)
            if result == "Draw":
                print("It's a draw!")
            else:
                print(f"Player {result} wins!")
            break

        turn += 1

if __name__ == "__main__":
    two_player_tic_tac_toe()
```

```python
import random

def get_computer_move(board):
    """Get a random move for the computer."""
    empty_squares = [(r, c) for r in range(3) for c in range(3) if board[r][c] == " "]
    return random.choice(empty_squares)

def player_vs_computer_tic_tac_toe():
    """Main function to play Tic-Tac-Toe against the computer."""
    board = [[" " for _ in range(3)] for _ in range(3)]
    players = ["X", "O"]
    human_player = input("Do you want to be X or O? ").upper()
    computer_player = "O" if human_player == "X" else "X"
    turn = 0

    while True:
        display_board(board)
        current_player = players[turn % 2]

        if current_player == human_player:
            print("Your turn.")
            while True:
                try:
                    row, col = map(int, input("Enter row and column (0-2) separated by a space: ").split())
                    if board[row][col] == " ":
                        board[row][col] = current_player
                        break
                    else:
                        print("That square is already taken. Try again.")
                except (ValueError, IndexError):
                    print("Invalid input. Enter row and column numbers between 0 and 2.")
        else:
            print("Computer's turn.")
            row, col = get_computer_move(board)
            board[row][col] = current_player

        # Check for a winner or draw
        result = check_winner(board)
        if result:
            display_board(board)
            if result == "Draw":
                print("It's a draw!")
            else:
                print(f"Player {result} wins!")
            break

        turn += 1

if __name__ == "__main__":
    player_vs_computer_tic_tac_toe()
```

```python
import math
import turtle

# Set up the screen
window = turtle.Screen()
window.bgcolor("black")

# Set up the turtle
pen = turtle.Turtle()
pen.speed(0)  # Fastest drawing speed
pen.color("red")

# Function to define the heart shape
def heart(t):
    x = 16 * math.sin(t) ** 3
    y = 13 * math.cos(t) - 5 * math.cos(2 * t) - 2 * math.cos(3 * t) - math.cos(4 * t)
    return x, y

# Draw the heart pattern
pen.penup()
for i in range(10):  # Reduce the number of layers for faster rendering
    pen.goto(0, 0)  # Reset position
    pen.pendown()
    for t in range(0, 100, 2):  # Increase the step size to reduce points
        x, y = heart(t / 10)
        pen.goto(x * i, y * i)
    pen.penup()

# Hide the turtle and keep the window open
pen.hideturtle()
turtle.done()
```

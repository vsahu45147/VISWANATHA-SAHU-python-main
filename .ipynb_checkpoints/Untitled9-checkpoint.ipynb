```python
import folium
from geopy.geocoders import Nominatim
import webbrowser
import os

# Ask the user for a location
location_name = input("Enter a location: ")

# Initialize the geocoder
geolocator = Nominatim(user_agent="geoapi")

# Get the coordinates of the location
location = geolocator.geocode(location_name)

if location:
    # Get latitude and longitude
    latitude = location.latitude
    longitude = location.longitude

    # Create a Folium map centered on the user's location
    my_map = folium.Map(location=[latitude, longitude], zoom_start=12)

    # Add a marker for the searched location
    folium.Marker([latitude, longitude], popup=location_name).add_to(my_map)

    # Save the map as an HTML file
    map_file = "searched_location_map.html"
    my_map.save(map_file)

    # Open the map in the default web browser
    webbrowser.open('file://' + os.path.realpath(map_file))

    print(f"✅ Map for '{location_name}' created successfully and opened in your browser!")
else:
    print("❌ Location not found. Please try again.")
```

```python
import pandas as pd

# Load original schedule
df = pd.read_csv("ipl_2025_schedule.csv")

# -------------------------
# Date cleanup
# -------------------------
df["Date"] = pd.to_datetime(df["Date"], format="%d %b %Y")

# -------------------------
# FACT TABLE
# -------------------------
fact_matches = df.copy()
fact_matches["Team 1"] = fact_matches["Fixture"].str.split(" vs ").str[0]
fact_matches["Team 2"] = fact_matches["Fixture"].str.split(" vs ").str[1]

fact_matches = fact_matches[[
    "Match No", "Date", "Team 1", "Team 2", "Venue", "Time (IST)"
]]

# -------------------------
# DIMENSION: TEAMS
# -------------------------
teams = pd.unique(
    fact_matches[["Team 1", "Team 2"]].values.ravel()
)
dim_teams = pd.DataFrame({
    "Team ID": range(1, len(teams) + 1),
    "Team Name": teams
})

# -------------------------
# DIMENSION: VENUES
# -------------------------
dim_venues = pd.DataFrame({
    "Venue ID": range(1, fact_matches["Venue"].nunique() + 1),
    "Venue Name": fact_matches["Venue"].unique()
})

# -------------------------
# DIMENSION: DATES
# -------------------------
dim_dates = pd.DataFrame({
    "Date": fact_matches["Date"].unique()
})
dim_dates["Year"] = dim_dates["Date"].dt.year
dim_dates["Month"] = dim_dates["Date"].dt.month
dim_dates["Month Name"] = dim_dates["Date"].dt.month_name()
dim_dates["Day"] = dim_dates["Date"].dt.day
dim_dates["Weekday"] = dim_dates["Date"].dt.day_name()

# -------------------------
# EXPORT
# -------------------------
fact_matches.to_csv("fact_matches.csv", index=False)
dim_teams.to_csv("dim_teams.csv", index=False)
dim_venues.to_csv("dim_venues.csv", index=False)
dim_dates.to_csv("dim_dates.csv", index=False)

print("✅ Power BI–ready files exported successfully!")
```
